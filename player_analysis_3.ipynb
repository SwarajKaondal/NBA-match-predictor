```python
import pandas as pd
import numpy as np
```

```python
playersDf = pd.read_csv("merged_players.csv")
teamsWithWinDf = pd.read_csv("teamsWithWin.csv")
```

```python
playersDf.shape
```

```python
teamsWithWinDf
```

```python
# Player id column is string for some reason, hence first convert into a list 
teamsWithWinDf.player_id = teamsWithWinDf.player_id.str[1:-1].str.split(',').tolist()
```

```python
# Now explode the dataframe
expandedTeamsWithWinDf = teamsWithWinDf.explode('player_id').reset_index()

expandedTeamsWithWinDf
```

```python
type(expandedTeamsWithWinDf["player_id"][0])
```

```python
playersDf["id"] = playersDf["id"].astype(str)
type(playersDf["id"][0])
```

```python
# Now join with the player stats
joinedTable = pd.merge(expandedTeamsWithWinDf, playersDf, left_on="player_id", right_on="id")
joinedTable
```
