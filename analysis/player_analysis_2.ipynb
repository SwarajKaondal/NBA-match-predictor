```python
import pandas as pd
import numpy as np
from fuzzywuzzy import fuzz
```

```python

teamsWithWinsDf = pd.read_csv("teamsWithWin.csv")
playerDf = pd.read_csv("Kaggle Data/player.csv")
playerOgDf = pd.read_csv("Data/players.txt")
teamsDf = pd.read_csv("Data/player_regular_season.txt")
kagglePlayerinfo = pd.read_csv("Kaggle Data/common_player_info.csv")
```

```python
kagglePlayerinfo["birth_year"] = kagglePlayerinfo.apply(lambda x: x["birthdate"][0:4], axis=1)
kagglePlayerinfo['first_name'] = kagglePlayerinfo['first_name'].str.strip().str.lower()
kagglePlayerinfo['last_name'] = kagglePlayerinfo['last_name'].str.strip().str.lower()
kagglePlayerinfo['person_id'].nunique()
```

```python
# convert og player data's firstname lastname to lowercase
playerOgDf['firstname'] = playerOgDf['firstname'].str.lower().str.strip()
playerOgDf['lastname'] = playerOgDf['lastname'].str.lower().str.strip()
playerOgDf = playerOgDf.rename(columns={'firstname': 'first_name'})
playerOgDf = playerOgDf.rename(columns={'lastname': 'last_name'})
playerOgDf["birth_year"] = playerOgDf.apply(lambda x: str(x["birthdate"])[0:4], axis=1)
playerOgDf['ilkid'].nunique()
```

```python
# convert kaggle player data's firstname lastname to lowercase
playerDf['first_name'] = playerDf['first_name'].str.strip().str.lower()
playerDf['last_name'] = playerDf['last_name'].str.strip().str.lower()
playerDf["id"].nunique()
```

```python
# merge both the dataframes
# mergedPlayers = pd.merge(playerOgDf, playerDf, left_on=['firstname', 'lastname'], right_on=['first_name','last_name'])
# mergedPlayers[mergedPlayers['last_name'] == "able"]
mergedPlayers = playerOgDf.merge(kagglePlayerinfo, on=['first_name', 'last_name','birth_year'])
print("Total rows = "+str(len(mergedPlayers.index)))
print("Unique ilkid = "+ str(mergedPlayers["ilkid"].nunique()))
print("Unique id = "+ str(mergedPlayers["person_id"].nunique()))

mergedPlayers
```
