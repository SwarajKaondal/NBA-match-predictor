```python
import pandas as pd
import numpy as np
```

```python

teamsWithWinsDf = pd.read_csv("teamsWithWin.csv")
playerDf = pd.read_csv("Kaggle Data/player.csv")
playerOgDf = pd.read_csv("Data/players.txt")
```

```python
# convert og player data's firstname lastname to lowercase
playerOgDf['firstname'] = playerOgDf['firstname'].str.lower()
playerOgDf['lastname'] = playerOgDf['lastname'].str.lower()
playerOgDf
```

```python
# convert kaggle player data's firstname lastname to lowercase
playerDf['first_name'] = playerDf['first_name'].str.lower()
playerDf['last_name'] = playerDf['last_name'].str.lower()
playerDf
```

```python
# merge both the dataframes
mergedPlayers = pd.merge(playerOgDf, playerDf, left_on=['firstname', 'lastname'], right_on=['first_name','last_name'])
mergedPlayers
```

```python
# Find out which players did not merge
merged_df = pd.merge(playerOgDf, mergedPlayers, on='ilkid', how='left', indicator=True)
rows_only_in_playerOfDf = merged_df[merged_df['_merge'] == 'left_only']
rows_only_in_playerOfDf

# Dont know what to do with this. Seems like height, weight, etc are NaNs for this
```
