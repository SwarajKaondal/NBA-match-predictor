# Creating normalized and cumilative player statistics
The player statistics in the CMU data are recorded for the entire season that is the points scored by a player in all the games that they played in that season, we are interested in knowing the statistics per game for a player. If we use the raw data for a season, players which have player more games will have an edge over the players which have played less games. Therefore we normalize these statistics by dividing these attributes by the number of games player each season. Once this is done, for every year we want an average of these statistics including the past years.

```python
import pandas as pd

df = pd.read_csv("player_with_game_id.csv")
gamesDf = pd.read_csv("Kaggle Data/game.csv")
```

```python
playerRegularSeason = pd.read_csv("Data/player_regular_season.txt")
```

```python
# Normalize the player stats by dividing stats by games played
playerRegularSeason1 = playerRegularSeason.iloc[:,:7]
playerRegularSeason2 = playerRegularSeason.iloc[:,7:]
playerRegularSeason2Nomarlized = playerRegularSeason2.div(playerRegularSeason1["gp"], axis=0).round(2)
playerRegularSeason2Nomarlized
```

```python
# Rejoin and form the full table
playerRegularSeasonNormalized = pd.concat([playerRegularSeason1, playerRegularSeason2Nomarlized], axis=1)
playerRegularSeasonNormalized
```

```python
playerRegularSeasonNormalized.to_csv("player_regular_season_normalized.csv")
```

```python
playerRegularSeasonNormalized = playerRegularSeasonNormalized[playerRegularSeasonNormalized["year"] >= 1995]
playerRegularSeasonNormalized
```

```python
# I want to calculate a cummalative table where a player stats for 1996 for ex is the stat of 1995
playerRegularSeasonNormalized.sort_values("year", inplace=True)

# Get all columns for cummalative sume
column_names = playerRegularSeasonNormalized.columns.to_list()
relevant_columns = column_names[6:]

# Find cummalative sum for each player for each stat
playerRegularSeasonNormalized['count'] = playerRegularSeasonNormalized.groupby('ilkid').cumcount() + 1
for col in relevant_columns:
    playerRegularSeasonNormalized[f'{col}_cumsum'] = playerRegularSeasonNormalized.groupby('ilkid')[col].cumsum()
    playerRegularSeasonNormalized[f'avg_{col}'] = playerRegularSeasonNormalized[f'{col}_cumsum'] / playerRegularSeasonNormalized['count']
playerRegularSeasonNormalized
```

```python
playerRegularSeasonNormalized.to_csv("cumilative_player_attributes.csv")
```
