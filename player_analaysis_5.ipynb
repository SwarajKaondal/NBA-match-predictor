```
import pandas as pd

# Example DataFrame
data = {
    'year': [2003, 2003, 2004, 2004, 2005, 2005],
    'player_id': ['abc', 'def', 'abc', 'def', 'abc', 'def'],
    'points': [10, 15, 20, 25, 30, 35]
}
df = pd.DataFrame(data)

# Sort DataFrame by 'year'
df.sort_values(by='year', inplace=True)

# Calculate average points scored prior to each year
df['avg_points_prior'] = (
    df.groupby('player_id')['points']
    .transform(lambda x: (x.shift(1) + x.shift(2)) / 2)
)

print(df)
```

````python
# I want to calculate a cummalative table where a player stats for 1996 for ex is the stat of 1995
playerNormalizedDf.sort_values("year", inplace=True)

# Get all columns for cummalative sume
column_names = playerNormalizedDf.columns.to_list()
relevant_columns = column_names[7:]

# Find cummalative sum for each player for each stat
playerNormalizedDf['count'] = playerNormalizedDf.groupby('ilkid').cumcount() + 1
for col in relevant_columns:
    playerNormalizedDf[f'{col}_cumsum'] = playerNormalizedDf.groupby('ilkid')[col].cumsum()
    playerNormalizedDf[f'avg_{col}'] = playerNormalizedDf[f'{col}_cumsum'] / playerNormalizedDf['count']    
````
